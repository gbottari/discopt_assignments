```python
def test_new_idea():
    problem = get_problem_by_filename('tsp_200_2')
    #problem = get_problem_by_filename('tsp_85900_1')
    solver = NewIdeaTSPSolver(2000000, alpha=0.999999, t0=300)
    #solver = Greedy2OptTSPSolver(400000)
    return solver._solve(problem), solver    
   
solution, solver = test_new_idea()
```

```python
def plot(solution, solver):
    x = solution.stats.improvements_x
    y = solution.stats.improvements_y
    print(solution.stats.final_value)
    
    fig = plt.figure()
    plt.plot(x, y)    
    #plt.xlim([2000, 15000])
    #plt.ylim([0, 60000])
    plt.grid()
    plt.title('obj')
    
    fig = plt.figure()
    plt.plot(list(range(len(solution.stats.temperature))), solution.stats.temperature)
    plt.grid()
    plt.title('temp')
    
    fig = plt.figure()
    plt.plot(list(range(len(solution.stats.probs))), solution.stats.probs, '.', alpha=0.3)
    plt.grid()
    plt.title('probs')
    
    #fig = plt.figure()
    #plt.scatter(x=[p.x for p in solution.problem.points], y=[p.y for p in solution.problem.points], alpha=0.5)
    
    plt.show()

plot(solution, solver)
```
